```python
from tkinter import *
from tkinter import ttk, scrolledtext, messagebox, filedialog
import random
import datetime

# Function to calculate total bills
def total_bills():
    # Drinks Items Price
    lassi_price = 50
    coffee_price = 20
    tea_price = 10
    juice_price = 30
    shakes_price = 50
    milk_price = 20
    shikanji_price = 15
    redbull_price = 150

    # Foods Items Price
    roti_price = 5
    dal_makhni_price = 120
    mutter_panner_price = 150
    paratha_price = 40
    mix_veg_price = 70
    omelete_price = 20
    veg_biryani_price = 120
    rice_price = 50

    # Drinks Item quantity
    lassi_q = int(lassi_qty.get()) if lassi_qty.get().isdigit() else 0
    coffee_q = int(coffee_qty.get()) if coffee_qty.get().isdigit() else 0
    tea_q = int(tea_qty.get()) if tea_qty.get().isdigit() else 0
    juice_q = int(juice_qty.get()) if juice_qty.get().isdigit() else 0
    shakes_q = int(shakes_qty.get()) if shakes_qty.get().isdigit() else 0
    milk_q = int(milk_qty.get()) if milk_qty.get().isdigit() else 0
    shikanji_q = int(shikanji_qty.get()) if shikanji_qty.get().isdigit() else 0
    redbull_q = int(redbull_qty.get()) if redbull_qty.get().isdigit() else 0

    # Foods Item quantity
    roti_q = int(roti_qty.get()) if roti_qty.get().isdigit() else 0
    dal_makhni_q = int(dal_makhni_qty.get()) if dal_makhni_qty.get().isdigit() else 0
    mutter_panner_q = int(mutter_panner_qty.get()) if mutter_panner_qty.get().isdigit() else 0
    paratha_q = int(paratha_qty.get()) if paratha_qty.get().isdigit() else 0
    mix_veg_q = int(mix_veg_qty.get()) if mix_veg_qty.get().isdigit() else 0
    omelete_q = int(omelete_qty.get()) if omelete_qty.get().isdigit() else 0
    veg_biryani_q = int(veg_biryani_qty.get()) if veg_biryani_qty.get().isdigit() else 0
    rice_q = int(rice_qty.get()) if rice_qty.get().isdigit() else 0

    # Calculate the total price for each item
    total_drinks_price = (lassi_q * lassi_price + coffee_q * coffee_price +
                          tea_q * tea_price + juice_q * juice_price +
                          shakes_q * shakes_price + milk_q * milk_price +
                          shikanji_q * shikanji_price + redbull_q * redbull_price)

    total_foods_price = (roti_q * roti_price + dal_makhni_q * dal_makhni_price +
                         mutter_panner_q * mutter_panner_price + paratha_q * paratha_price +
                         mix_veg_q * mix_veg_price + omelete_q * omelete_price +
                         veg_biryani_q * veg_biryani_price + rice_q * rice_price)

    # Calculate overall total
    overall_total = total_drinks_price + total_foods_price

    # Display the result
    cost_of_drinks_var.set(f"{total_drinks_price} ₹")
    cost_of_foods_var.set(f"{total_foods_price} ₹")
    sub_total_var.set(f"{overall_total} ₹")
    paid_tax = overall_total * 0.05
    paid_tax_var.set(f"{paid_tax:.2f} ₹")
    total_cost_var.set(f"{overall_total + paid_tax:.2f} ₹")

# GUI Setup
root = Tk()
root.title("Restaurant Management System")
root.geometry("800x300")

# Variables
lassi_qty = StringVar()
coffee_qty = StringVar()
tea_qty = StringVar()
juice_qty = StringVar()
shakes_qty = StringVar()
milk_qty = StringVar()
shikanji_qty = StringVar()
redbull_qty = StringVar()

roti_qty = StringVar()
dal_makhni_qty = StringVar()
mutter_panner_qty = StringVar()
paratha_qty = StringVar()
mix_veg_qty = StringVar()
omelete_qty = StringVar()
veg_biryani_qty = StringVar()
rice_qty = StringVar()

cost_of_drinks_var = StringVar()
cost_of_foods_var = StringVar()
sub_total_var = StringVar()
paid_tax_var = StringVar()
total_cost_var = StringVar()

# GUI Layout
Label(root, text="Restaurant Management System", font=("Arial", 20)).pack()

# Drinks Frame
drinks_frame = LabelFrame(root, text="Drinks", font=("Arial", 15))
drinks_frame.pack(side=LEFT, fill="both", expand="yes")

# Drinks Items
Label(drinks_frame, text="Lassi").grid(row=0, column=0)
Entry(drinks_frame, textvariable=lassi_qty).grid(row=0, column=1)

Label(drinks_frame, text="Coffee").grid(row=1, column=0)
Entry(drinks_frame, textvariable=coffee_qty).grid(row=1, column=1)

Label(drinks_frame, text="Tea").grid(row=2, column=0)
Entry(drinks_frame, textvariable=tea_qty).grid(row=2, column=1)

Label(drinks_frame, text="Juice").grid(row=3, column=0)
Entry(drinks_frame, textvariable=juice_qty).grid(row=3, column=1)

Label(drinks_frame, text="Shakes").grid(row=4, column=0)
Entry(drinks_frame, textvariable=shakes_qty).grid(row=4, column=1)

Label(drinks_frame, text="Milk").grid(row=5, column=0)
Entry(drinks_frame, textvariable=milk_qty).grid(row=5, column=1)

Label(drinks_frame, text="Shikanji").grid(row=6, column=0)
Entry(drinks_frame, textvariable=shikanji_qty).grid(row=6, column=1)

Label(drinks_frame, text="Redbull").grid(row=7, column=0)
Entry(drinks_frame, textvariable=redbull_qty).grid(row=7, column=1)

# Foods Frame
foods_frame = LabelFrame(root, text="Foods", font=("Arial", 15))
foods_frame.pack(side=LEFT, fill="both", expand="yes")

# Foods Items
Label(foods_frame, text="Roti").grid(row=0, column=0)
Entry(foods_frame, textvariable=roti_qty).grid(row=0, column=1)

Label(foods_frame, text="Dal Makhni").grid(row=1, column=0)
Entry(foods_frame, textvariable=dal_makhni_qty).grid(row=1, column=1)

Label(foods_frame, text="Mutter Panner").grid(row=2, column=0)
Entry(foods_frame, textvariable=mutter_panner_qty).grid(row=2, column=1)

Label(foods_frame, text="Paratha").grid(row=3, column=0)
Entry(foods_frame, textvariable=paratha_qty).grid(row=3, column=1)

Label(foods_frame, text="Mix Veg").grid(row=4, column=0)
Entry(foods_frame, textvariable=mix_veg_qty).grid(row=4, column=1)

Label(foods_frame, text="Omelete").grid(row=5, column=0)
Entry(foods_frame, textvariable=omelete_qty).grid(row=5, column=1)

Label(foods_frame, text="Veg Biryani").grid(row=6, column=0)
Entry(foods_frame, textvariable=veg_biryani_qty).grid(row=6, column=1)

Label(foods_frame, text="Rice").grid(row=7, column=0)
Entry(foods_frame, textvariable=rice_qty).grid(row=7, column=1)

# Bill Frame
bill_frame = LabelFrame(root, text="Bill", font=("Arial", 15))
bill_frame.pack(side=RIGHT, fill="both", expand="yes")

# Bill Items
Label(bill_frame, text="Cost of Drinks").grid(row=0, column=0)
Entry(bill_frame, textvariable=cost_of_drinks_var).grid(row=0, column=1)

Label(bill_frame, text="Cost of Foods").grid(row=1, column=0)
Entry(bill_frame, textvariable=cost_of_foods_var).grid(row=1, column=1)

Label(bill_frame, text="Sub Total").grid(row=2, column=0)
Entry(bill_frame, textvariable=sub_total_var).grid(row=2, column=1)

Label(bill_frame, text="Paid Tax").grid(row=3, column=0)
Entry(bill_frame, textvariable=paid_tax_var).grid(row=3, column=1)

Label(bill_frame, text="Total Cost").grid(row=4, column=0)
Entry(bill_frame, textvariable=total_cost_var).grid(row=4, column=1)

# Buttons
Button(root, text="Total", command=total_bills).pack(side=LEFT)
Button(root, text="Save", command=lambda: print("Save functionality not implemented")).pack(side=LEFT)
Button(root, text="Send", command=lambda: print("Send functionality not implemented")).pack(side=LEFT)
Button(root, text="Exit", command=root.quit).pack(side=RIGHT)

# Run the application
root.mainloop()
```
